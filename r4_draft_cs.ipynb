```python
#!/usr/bin/env python
# coding: utf-8

import csv
import pandas as pd


r4_l234=pd.read_csv("./raw/draft/R4_CS_draft_L234.csv")
h28_working=pd.read_csv("./raw/draft/H28_CS_draft.csv")

h28_working
r4_l4_from_h28=h28_working.loc[:,['第3層','第4層','H28対応項目','メモ']].dropna(subset=['第3層'])
r4_l4=pd.merge(r4_l234.loc[:,["第2層","第3層"]],r4_l4_from_h28,how="outer",on="第3層")
r4_l4.to_csv("./dist/draft_cs/l4.csv",encoding="utf_8_sig",index="False")

symptoms=h28_working.loc[:,['症候','第4層','H28対応項目','メモ']].dropna(subset=['症候']).fillna("")
symptoms=symptoms.groupby(['症候'],sort=False)\
            .apply(lambda x:pd.DataFrame({
                "H28対応項目":[",".join(list(x.H28対応項目))],
                "第4層":["\n".join(list(x.第4層))],
                "メモ":["\n".join(list(x.メモ))],
                })).reset_index()
symptoms.to_csv("./dist/draft_cs/symptoms.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
symptoms
```
